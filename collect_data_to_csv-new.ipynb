```python
import serial
import csv
import time

# Set up the serial connection
ser = serial.Serial('COM3', 9600)  # Replace 'COM_PORT' with your port, e.g., 'COM3' or '/dev/ttyACM0'
time.sleep(2)  # Wait for the serial connection to initialize

# Open a CSV file to save the data
with open('sensor_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    # Header row
    writer.writerow(['Timestamp', 'Gyro_X', 'Gyro_Y', 'Gyro_Z', 'Accel_X', 'Accel_Y', 'Accel_Z'])  

    while True:
        if ser.in_waiting > 0:  # Check if there's data in the serial buffer
            data = ser.readline().decode('utf-8').strip()  # Read a line of data
            # Split the data by commas (assuming it's in CSV format)
            values = data.split(',')
            
            if len(values) == 7:  # Check that we have the correct number of values (timestamp, x, y, z)
                writer.writerow(values)  # Write the data to the CSV file
                print(data)  # Optionally print it to the console
```
